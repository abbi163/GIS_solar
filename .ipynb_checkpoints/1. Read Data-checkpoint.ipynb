```python
import pandas as pd
import numpy as np
```

### Reading Global PV potential data !!

```python
# Reading xlxs data !!
file='DATA/Global_Photovoltaic_Potential_Country_Rankings.xlsx'
sheet = 'Country indicators'
dfs = pd.read_excel(file, sheet_name=sheet, skiprows=1)
```

```python
dfs.head()
```

```python
dfs.columns
```

```python
# extracting ISO_A3 , Average Theoratical potential and Average practical potential data !!
temp1 = dfs[['ISO_A3', 'Average theoretical potential (GHI, kWh/m2/day), \nlong-term',
    'Average practical potential \n(PVOUT Level 1, \nkWh/kWp/day), long-term']]

```

```python
temp1.head()
```

```python
# Reading ISO_A3 file and merging it with the original file !

ISO_A3 = pd.read_csv('ISO_A3.csv')

for i in range(len(ISO_A3)):
    ISO_A3.iloc[i, 2] = ISO_A3.iloc[i,2].replace('"', '').replace(" ", '')
    ISO_A3.iloc[i, 4] = float(ISO_A3.iloc[i,4].replace('"', '').replace(" ", ''))
    ISO_A3.iloc[i, 5] = float(ISO_A3.iloc[i,5].replace('"', '').replace(" ", ''))
    
    
```

```python
ISO_A3.head()
```

```python
# Merging both together !!

temp_2 = temp1.merge(ISO_A3, left_on='ISO_A3', right_on='Alpha-3 code')
temp_2.dropna()
temp_2 = temp_2[['ISO_A3', 'Country', 'Latitude (average)', 'Longitude (average)', 'Average theoretical potential (GHI, kWh/m2/day), \nlong-term',
       'Average practical potential \n(PVOUT Level 1, \nkWh/kWp/day), long-term']]
temp_2.to_csv('temp_2.csv', index=None)
```
